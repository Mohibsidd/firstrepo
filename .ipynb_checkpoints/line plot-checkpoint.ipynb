```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean  
```

```python
iris=sns.load_dataset("iris")
```

### adding title

```python
sns.lineplot(data=iris)
plt.title('flower graph')
```

```python
sns.lineplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.lineplot(x='sepal_length',y= 'petal_width',data=iris)
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=iris)
```

## adding limits 

```python
sns.lineplot(data=iris)
plt.title('flower graph')
plt.xlim(1)
plt.ylim(2)
```

## size of figuers

```python
sns.lineplot(data=iris)
plt.figure(figsize=(8,16))
```

# bar plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
# load dataset
phool=sns.load_dataset('iris')
phool.head()
```

```python
sns.barplot(data=phool)
```

```python
sns.barplot(x="species",y="sepal_length",data=phool)
```

```python
sns.barplot(x="species",y="petal_length",data=phool)
```

```python
ship=sns.load_dataset('titanic')
ship.head()
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=ship,order=['male','female'],ci=None)
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=ship,order=['male','female'],ci=None
           ,palette='pastel')
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=ship,order=['male','female'],ci=None,palette='Greens')
```

```python
sns.barplot(x='class',y='fare',hue='sex',data=ship,estimator=mean,palette='Greens',saturation=.1)
```
